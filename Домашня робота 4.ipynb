## 3 Різні параметри

### 3.1 Біноміальний розподіл

Зобразіть графічно функцію щільності, функцію розподілу та гістограму $10000$ випадкових величин біноміального розподілу з параметрами:
1. $n = 15, p = 0.8$
2. $n = 15, p = 0.2$
3. $n = 20, p = 0.4$
4. $n = 20, p = 0.6$
5. $n = 7, p = 0.5$
6. $n = 25, p = 0.5$
7. $n = 100, p = 0.5$

```python
from scipy.stats import binom
def graphic(n, p):
    rv = binom.rvs(n, p, size = 10000)
    uniq_rv = np.unique(rv)
    counts_ = []
    for i in uniq_rv:
        counts_.append(sum(rv == i))
        print("count of {} is: {}".format(i, sum(rv == i)))
    y1 = binom.pmf(uniq_rv, n, p)
    plt.plot(uniq_rv, y1)
    plt.show()
    y2 = binom.cdf(uniq_rv, n, p)
    plt.plot(uniq_rv, y2)
    plt.show()
    plt.hist(rv)
    plt.show()
questions = [[15, 0.8], [15, 0.2], [20, 0.4], [20, 0.6], [7, 0.5], [25, 0.5], [100, 0.5]]
for n, p in questions:
    print("n = {}, p = {}".format(n, p))
    graphic(n, p)

```

## 4 Апроксимація біноміального розподілу розподілом Пуассона

Покажіть схожість біноміального розподілу та розподілу Пуассона (їх функцій щільності) з відповідними параметрами.
Наприклад:
- $n = 100, p = 0.05$
- $n = 100, p = 0.1$
- $n = 100, p = 0.15$
- $n = 100, p = 0.2$
- $n = 100, p = 0.35$
- $n = 100, p = 0.5$
- $n = 100, p = 0.75$

```python
from scipy.stats import binom, poisson
import numpy as np
def graphic(n, p):
    rv1 = binom.rvs(n, p)
    rv2 = poisson.rvs(n, p)
    uniq_rv1 = np.unique(rv1)
    uniq_rv2 = np.unique(rv2)
    print("binom variables:")
    counts_1 = []
    for i in uniq_rv1:
        counts_1.append(sum(rv1 == i))
        print("count of {} is: {}".format(i, sum(rv1 == i)))
    print("poisson variables")
    counts_2 = []
    for i in uniq_rv2:
        counts_2.append(sum(rv2 == i))
        print("count of {} is: {}".format(i, sum(rv2 == i)))
    y1 = binom.pmf(uniq_rv1, n, p)
    plt.plot(uniq_rv1, y1, label = "Binomial")
    y2 = binom.pmf(uniq_rv2, n, p)
    plt.plot(uniq_rv2, y2, label = "Poisson")
    plt.legend()
    plt.show()
questions = [[100, 0.05], [100, 0.1], [100, 0.15], [100, 0.2], [100, 0.35], [100, 0.5], [100, 0.75]]
for n, p in questions:
    print("n = {}, p = {}".format(n, p))
    graphic(n, p)
"""
Як я зрозумів, при більшому параметрі p, розподіли бінома та пуасона наче накладаються один на одного. 
Розподіл пуассона є продовженням розподіла бінома. Це видно при збільшенні p. 
"""
```

```python
from scipy.stats import binom, poisson
import numpy as np
import matplotlib.pyplot as plt
def graphic(n, p):
    x = np.arange(0, 100)
    y1 = binom.pmf(x, n, p)
    plt.plot(x, y1, label = "Binomial")
    y2 = poisson.pmf(x, n, p)
    plt.plot(x, y2, label = "Poisson")
    plt.legend()
    plt.show()
questions = [[100, 0.05], [100, 0.1], [100, 0.15], [100, 0.2], [100, 0.35], [100, 0.5], [100, 0.75]]
for n, p in questions:
    print("n = {}, p = {}".format(n, p))
    graphic(n, p)
```

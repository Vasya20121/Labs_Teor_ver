## 1 Пароль
Напишіть програму (функцію) для генерування випадкового пароля з такими обмеженнями:
- пароль складається з рівно $10$ символів
- пароль містить хоча б одну маленьку латинську літеру [a-z]
- пароль містить хоча б одну велику латинську літеру [A-Z]
- пароль містить хоча б одну цифру [0-9]
- пароль містить хоча б один спеціальний символ [~!@#$%^&*()-_+=\/{}[].,?<>:;]

```python
import numpy as np
from numpy.random import choice, shuffle
symbols = list('1234567890abcdefghigklmnopqrstuvyxwzABCDEFGHIGKLMNOPQRSTUVYXWZ!@#$%^&*()-=_|/?><~:;{}[]')
digits = list('1234567890')
lower_letters = list('abcdefghigklmnopqrstuvyxwz')
upper_letters = list('ABCDEFGHIGKLMNOPQRSTUVYXWZ')
spec_symbols = list('!@#$%^&*()-=_|/?><~:;{}[]')
shuffle(symbols)
def ran_pass(a = 1, b = 1, c = 1, d = 1, sym = 7, n = 10):
    passw = ''
    A = ''.join(choice(lower_letters, a)) # а - к-сть символів нижнього регістру, які вибираємо випадково
    B = ''.join(choice(upper_letters, b)) # і так далі
    C = ''.join(choice(spec_symbols, c)) #
    D = ''.join(choice(digits, d)) #
    S = ''.join(choice(symbols, sym)) #
    passw = list(A+B+C+D+S)
    shuffle(passw)
    passw = ''.join(passw)
    return passw
#Можна задати параметри, щоб поміняти умову задачі, якщо потрібно, але уважно
print("The random password is:", ran_pass())
    
```

- $k = 2, n = 10, 50, 100, 500, 1000, 5000, 10000$

## 4 Сума чисел
Реалізуйте функцію підрахунку точної (теоретичної) імовірності того, що сума чисел після $n$ кидків звичайного грального кубика дорівнює $s$, для таких параметрів:
- $n = 3, s = 13$
- $n = 2, s = 5$
- $n = 3, s = 7$

Тобто, якщо $X_i, i = 1, \ldots, n$, - число на $i$-му кидку кубика, а $X = \sum\limits_{i = 1}^{n} X_i$, то потрібно знайти ймовірність $\textbf{P} \left( X = s \right)$.

Підрахуйте також емпіричні ймовірності відповідних подій у випадку $10000$ симуляцій.

**Зауваження.** Якщо введені параметри не мають сенсу, то повертати $0$.
Наприклад, якщо при $n = 3$ вводиться $s < 3$ або $s > 18$.

```python

def task_4(n, s, all = True):
    try:
        k = 0
        num = choice([1,2,3,4,5,6], n)
        if (n < 1) or (n == 1 and (s < 1 or s > 6)) or (n == 2 and (s < 1 or s > 12)) or (n == 3 and (s < 1 or s > 18)):
            return "0"
        summ = sum(num)
        if all == True:
            print("n = ", n, "s = ", s)
            print(num)
            print(k)
            print("sum = ", summ)
            print()
        else:
            return summ
    except Exception:
        print("0")
task_4(3, 6)
task_4(2, 3)
task_4(3, 5)
questions = [(3, 13), (2, 5), (3, 7)]
for n, s in questions:
    prob = 0 
    emp_prob = 0
    for i in range(10000):
        if task_4(n, s, False) == s:
            prob += 1
    emp_prob = prob/10000
    print("tests = 10000, n = {}, s = {}".format(n, s))
    print(emp_prob)
task_4(0, 5)


```
